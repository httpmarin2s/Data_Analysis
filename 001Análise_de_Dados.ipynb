## **Observação geral**
**Chefe:** Qual o faturamento total?

```python
#Verificando o faturamento total
faturamento_total = df['Valor Final'].sum()
print(faturamento_total)
```

**Chefe:** Qual o faturamento total por loja? E qual loja teve o maior faturamento?

```python
#faturamento por loja
faturamento_por_loja = df[['ID Loja','Valor Final']].groupby('ID Loja').sum()
display(faturamento_por_loja)
```

**Chefe:** Quanto rendeu cada produto

```python
faturamento_por_produto = df[['ID Loja', 'Produto', 'Valor Final']].groupby(['ID Loja', 'Produto']).sum()
faturamento_por_produto = faturamento_por_produto.sort_values(by=['ID Loja', 'Valor Final'], ascending=[True, False])
display(faturamento_por_produto)
```

## **Desempenho das Lojas:**

**Chefe:** Qual loja apresentou o maior faturamento total no período analisado? E a menor?

```python
#Organizando do maior para o menor
faturamento_por_loja = faturamento_por_loja.sort_values(by='Valor Final', ascending=False)
display(faturamento_por_loja)
```



1.   **Maior Faturamento:** Iguatemi Campinas
2.   **Menor Faturamento:** Iguatemi Esplanada



**Chefe:** Qual loja teve o maior número de produtos vendidos?

```python
# Agrupar por loja e produto e calcular a soma da quantidade vendida
qtd_por_produto = df[['ID Loja', 'Produto', 'Quantidade']].groupby(['ID Loja', 'Produto']).sum()

# Ordenar a quantidade vendida por loja e, dentro de cada loja, por quantidade em ordem decrescente
qtd_por_produto = qtd_por_produto.sort_values(by=['ID Loja', 'Quantidade'], ascending=[True, False])

display(qtd_por_produto)
```

```python
# Calcular a soma total de produtos vendidos por loja
total_por_loja = df[['ID Loja', 'Quantidade']].groupby('ID Loja').sum()

# Ordenar a soma total de produtos vendidos por loja em ordem decrescente
total_por_loja = total_por_loja.sort_values(by='Quantidade', ascending=False)

display(total_por_loja)
```

```python
# Create a bar chart using Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(total_por_loja.index, total_por_loja['Quantidade'])
plt.xlabel('ID Loja')
plt.ylabel('Total de Produtos Vendidos')
plt.title('Total de Produtos Vendidos por Loja')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()

plt.show()
```
